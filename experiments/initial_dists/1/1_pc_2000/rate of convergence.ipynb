```python
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('1_vs_3.csv')
data = df.loc[df['seed']==23].to_numpy()
f = lambda x, a, b: a*np.exp(-b*x)
t = data[:, 0][:6]
y = data[:, -1][:6]
a, b = curve_fit(f, t, y)[0]
fig = plt.figure()
plt.plot(t, y, label='data')
plt.plot(t, f(t, a, b), label='fitted, a={:.2f}, b={:.2f}'.format(a, b))
plt.legend()
```
