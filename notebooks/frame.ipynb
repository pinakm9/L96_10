```python
import numpy as np
import scipy 
import pandas as pd 
import matplotlib.pyplot as plt
import tables 


class Plotter:
    def __init__(self, asml_file):
        self.file = asml_file 

    def plot(self, saveas, frame):
        fig = plt.figure(figsize=(8, 8))
        ax = fig.add_subplot(111)
        h5 = tables.open_file(self.file, mode='r')
        X = np.array(getattr(h5.root.particles, 'time_' + str(frame)).read().tolist())
        ax.scatter(X[:, 0], X[:, 1], label='PF')
        if saveas is not None:
            ax.savefig(saveas)
        plt.show()
```

```python
asml_file = '../effect of obs cov/data/prior_1_obs_gap_0.05_obs_cov_0.2_seed_3/assimilation.h5'
pltr = Plotter(asml_file).plot(None, int(4.0/0.05))
```
